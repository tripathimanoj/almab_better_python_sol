```python
# male famale sep ==>

def sex_sep(x):

    if x=='female':
        return 0
    else:
        return 1
    
data1['sex_no']=data1['Sex'].apply(sex_sep)
data1
```

```python
# fare group== >

def fare_group(x):
    if x <= 100 :
        return "A"
    elif x >100 and x<200:
        return "B"
    else :
        return "C"
    
data1['Fare_group'] = data1['Fare'].apply(fare_group)
data1
```
